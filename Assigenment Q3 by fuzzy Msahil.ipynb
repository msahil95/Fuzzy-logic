Write a fuzzy logic python program for a retail loan evaluation. The fuzzy model must consist of three input variables (‘Income Level’, ‘Credit History’, ‘Employment’), and a single output variable that indicates ‘Credit Standing’. Credit standing shall be regarded as ‘low’, ‘medium’ or ‘high’ depending on the degree of membership of the linguistic terms. The linguistic terms and the variable values are provided in the table below:


Variables
Linguistic terms
Variable values

Credit History

Bad/Average/Good

0–1

Employment

Short/Medium/Long

0–25

Income Level

Low/Medium/High

0–3,000

Credit Standing

Low/Medium/High

0–1



The rules are defined as follows:

Rule 1: If credit history is ‘Bad’, then credit standing is ‘Low’.

Rule 2: If credit history is ‘Good’, then credit standing is ‘High’.

Rule 3: If credit history is ‘Average’, then credit standing is ‘Medium’.

Rule 4: If credit history is “Bad” and employment period is ‘Short’ and income level is ‘Low’, then credit standing is ‘Low’.

Rule 5: If credit history is ‘Good’ and employment period is ‘Long’ and income level is ‘Medium’, then Credit standing is ‘Medium’.

Rule 6: If credit history is ‘Good’ and employment period is ‘Long’ and income level is 'High', then Credit standing is ‘High’.

```python
import numpy as np
from IPython.display import Image
from IPython.core.display import HTML 
import skfuzzy as fuzz
from skfuzzy import control as ctrl
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# New Antecedent/Consequent objects hold Variables ,Linguistic terms and Variable values
# functions
# Income Level= InL, Credit History= CreH, Employment= Empl and Credit Standing= CreS short cut used while coding   
InL = ctrl.Antecedent(np.arange(0, 3025, 1),'Income Level')
CreH = ctrl.Antecedent(np.arange(0, 1, 0.1),'Credit History')
Empl = ctrl.Antecedent(np.arange(0, 25, 1),'Employment')
CreS = ctrl.Consequent(np.arange(0, 1, 0.1),'Credit Standing')
```

```python
# Custom membership functions can be built interactively with a familiar,
# Pythonic API
CreS['low'] = fuzz.trimf(CreS.universe, [0, 0, 0.1])
CreS['medium'] = fuzz.trimf(CreS.universe, [0, 0.1, 0.5])
CreS['high'] = fuzz.trimf(CreS.universe, [0.1, 0.5, 1])
```

```python
InL['low'] = fuzz.trimf(InL.universe, [0, 0, 1000])
InL['medium'] = fuzz.trimf(InL.universe, [0, 1000, 1500])
InL['high'] = fuzz.trimf(InL.universe, [1000, 1500, 3000])

CreH['Bad'] = fuzz.trimf(CreH.universe, [0, 0, 0.1])
CreH['Average'] = fuzz.trimf(CreH.universe, [0, 0.1, 0.5])
CreH['Good'] = fuzz.trimf(CreH.universe, [0.1, 0.5, 1])

Empl['Short'] = fuzz.trimf(Empl.universe, [0, 0, 7])
Empl['Medium'] = fuzz.trimf(Empl.universe, [0, 7, 15])
Empl['Long'] = fuzz.trimf(Empl.universe, [7,15 , 25])

```

```python
#Rules 
rule1 = ctrl.Rule(CreH['Bad'] , CreS['low'])
rule2 = ctrl.Rule(CreH['Good'], CreS['high'])
rule3 = ctrl.Rule(CreH['Average'] , CreS['medium'])
rule4 = ctrl.Rule(CreH['Bad']&Empl['Short']&InL['low'] , CreS['low'])
rule5 = ctrl.Rule(CreH['Average']&Empl['Medium']&InL['medium'] , CreS['medium'])
rule6 = ctrl.Rule(CreH['Good']&Empl['Long']&InL['high'], CreS['high'])
```

```python
rule1.view()
rule4.view()
```

```python
CreSt_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6])
CreSt = ctrl.ControlSystemSimulation(CreSt_ctrl)
```

```python
rule1.view()
rule2.view()
rule3.view()
rule4.view()
rule5.view()
rule6.view()
```

```python
CreSt.input['Credit History'] = 0.4
CreSt.input['Employment'] = 17
CreSt.input['Income Level'] =1750
CreSt.compute()
print(CreSt.output['Credit Standing'])
```

```python
CreS.view(sim=CreSt)
```
